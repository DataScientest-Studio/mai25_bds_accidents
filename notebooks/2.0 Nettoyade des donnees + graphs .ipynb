```python
accidents_copy = accidents.copy()
print("Colonnes non numériques :",accidents_copy.select_dtypes(include=['object']).columns)
```

```python
colonnes_a_convertir = ['lat', 'long', 'voie', 'v2', 'nbv', 'pr', 'pr1', 'lartpc', 'larrout']
accidents_copy[colonnes_a_convertir] = accidents_copy[colonnes_a_convertir].apply(pd.to_numeric, errors='coerce')
```

```python
print(accidents_copy[colonnes_a_convertir].dtypes)
```

```python
print(accidents_copy.isnull().sum())
```

```python
accidents_copy.drop(columns=['occutc', 'lartpc', 'v2', 'lat', 'long'], inplace=True)
```

```python
print(accidents_copy.info())
```

```python
print(accidents_copy.isnull().sum())
```

```python
colonnes_a_garder = ['atm', 'lum', 'col', 'catv', 'sexe', 'an_nais', 'catu', 'agg', 'catr', 'circ', 'jour', 'mois', 'hrmn', 'grav']
accidents_final = accidents_copy[colonnes_a_garder]
```

```python
print(accidents_final.dtypes)
```

```python
accidents_final.loc[:, 'hrmn'] = pd.to_datetime(accidents_final['hrmn'], format='%H:%M')
accidents_final.loc[:, 'heure'] = accidents_final['hrmn'].dt.hour
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
accidents_corr = accidents_final.select_dtypes(include=['number'])
plt.figure(figsize=(12, 8))
sns.heatmap(accidents_corr.corr(), cmap="coolwarm", annot=True, linewidths=0.5)
plt.title("Heatmap des Corrélations entre les Variables")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=accidents_final['grav'], y=accidents_final['an_nais'], palette="coolwarm")
plt.title("Influence de l'Âge sur la Gravité des Accidents")
plt.xlabel("Gravité des Accidents (1 = léger, 4 = très grave)")
plt.ylabel("Année de Naissance")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=accidents_final['atm'], y=accidents_final['grav'], palette="viridis")
plt.title("Impact des Conditions Météo sur la Gravité des Accidents")
plt.xlabel("Météo")
plt.ylabel("Gravité Moyenne")
plt.show()
```

```python
accidents_final = accidents_final[(accidents_final['grav'] >= 1) & (accidents_final['grav'] <= 4)]
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=accidents_final['catv'], y=accidents_final['grav'], palette="Blues")
plt.title("Impact du Type de Véhicule sur la Gravité des Accidents")
plt.xlabel("Type de Véhicule")
plt.ylabel("Gravité Moyenne")
plt.show()
```
