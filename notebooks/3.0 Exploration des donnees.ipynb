```python
grav_mapping = {1: "Indemne",2: "Tué", 3: "Blessé hospitalisé", 4: "Blessé léger"}
accidents['grav_label'] = accidents['grav'].map(grav_mapping)
grav_counts = accidents['grav_label'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=grav_counts.index, y=grav_counts.values, palette='Reds')
plt.title("Répartition des accidents selon la gravité")
plt.xlabel("Gravité")
plt.ylabel("Nombre d'accidents")
plt.grid(axis='y')
plt.savefig("reports/figures/Repartition des accidents selon la gravite.png")
plt.show()
```

```python
sexe_mapping = {1: "Homme", 2: "Femme",-1: "Non renseigné"}
accidents['sexe_label'] = accidents['sexe'].map(sexe_mapping)
sexe_counts = accidents['sexe_label'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=sexe_counts.index, y=sexe_counts.values, palette='Set2')
plt.title("Répartition des accidents selon le sexe")
plt.xlabel("Sexe")
plt.ylabel("Nombre d'accidents")
plt.grid(axis='y')
plt.savefig("reports/figures/Repartition des accidents selon le sexe.png")
plt.show()
```

```python
accidents['dep'] = accidents['dep'].astype(str).str.strip()
dep_counts = accidents['dep'].value_counts().sort_values(ascending=False).head(20)  
plt.figure(figsize=(12, 6))
sns.barplot(x=dep_counts.index, y=dep_counts.values, palette='coolwarm')
plt.title("Top 20 des départements avec le plus d'accidents")
plt.xlabel("Département")
plt.ylabel("Nombre d'accidents")
plt.grid(axis='y')
plt.savefig("reports/figures/Repartition des accidents par departement.png")
plt.show()
```

```python
df = accidents.copy()
df = df[(df['age'] > 0) & (df['age'] < 100)]
df['tranche_age'] = pd.cut(df['age'], bins=range(0, 101, 5))
age_sex = df.groupby(['tranche_age', 'sexe_label']).size().unstack().fillna(0)
age_sex['Homme'] = -age_sex['Homme']
plt.figure(figsize=(10, 8))
age_sex[['Homme', 'Femme']].plot(kind='barh', stacked=False, color=['skyblue', 'lightcoral'])
plt.title("Pyramide des âges selon le sexe")
plt.xlabel("Nombre d'accidents")
plt.ylabel("Tranches d’âge")
plt.axvline(0, color='black')
plt.grid(axis='x')
plt.legend(loc='upper right')
plt.tight_layout()
plt.savefig("reports/figures/Pyramide_age_sexe_simple.png")
plt.show()
```

```python
df = accidents.copy()
df = df[(df['age'] > 0) & (df['age'] < 100)]
grav_mapping = {1: "Indemne", 2: "Tué", 3: "Blessé hospitalisé", 4: "Blessé léger"}
df['grav_label'] = df['grav'].map(grav_mapping)
age_grav = df.groupby(['age', 'grav_label']).size().reset_index(name='count')
plt.figure(figsize=(12, 6))
sns.lineplot(data=age_grav, x='age', y='count', hue='grav_label', palette='Set1')
plt.title("Gravité des accidents selon l’âge")
plt.xlabel("Âge")
plt.ylabel("Nombre d'accidents")
plt.grid(True)
plt.legend(title="Gravité")
plt.tight_layout()
plt.savefig("reports/figures/Gravite_des_accidents_selon_age.png")
plt.show()
```

```python
df = accidents.copy()
grav_mapping = {1: "Indemne", 2: "Tué", 3: "Blessé hospitalisé", 4: "Blessé léger"}
df['grav_label'] = df['grav'].map(grav_mapping)
grav_sexe_counts = df.groupby(['sexe_label', 'grav_label']).size().reset_index(name='count')
total_by_sex = grav_sexe_counts.groupby('sexe_label')['count'].transform('sum')
grav_sexe_counts['percent'] = grav_sexe_counts['count'] / total_by_sex * 100
plt.figure(figsize=(10,6))
sns.barplot(data=grav_sexe_counts, x='sexe_label', y='percent', hue='grav_label', palette='Set2')
plt.title("Répartition de la gravité des accidents selon le sexe")
plt.xlabel("Sexe")
plt.ylabel("Pourcentage d'accidents (%)")
plt.legend(title="Gravité")
plt.grid(axis='y')
plt.tight_layout()
plt.savefig("reports/figures/Gravite_accidents_selon_sexe.png")
plt.show()
```

```python
collision_counts = accidents['col'].value_counts().sort_index()
col_mapping = {1: "Frontal", 2: "Par l'arrière", 3: "Latéral",4: "En chaine", 5: "Multiples", 6: "Autres", 7: "Sans collision"}
collision_labels = [col_mapping.get(i, f"Type {i}") for i in collision_counts.index]
plt.figure(figsize=(10,6))
sns.barplot(x=collision_labels, y=collision_counts.values, palette="pastel")
plt.title("Répartition des types de collisions")
plt.xlabel("Type de collision")
plt.ylabel("Nombre d'accidents")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.savefig("reports/figures/Analyse_des_collisions.png")
plt.show()
```
